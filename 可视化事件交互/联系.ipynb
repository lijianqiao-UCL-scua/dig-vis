```python
import pandas as pd
```

```python
od = pd.read_csv(r'职住OD.csv')
od
```

```python
od1 = od[['居住地_市','就业地_市','count']]
```

```python
od1['width'] = od1['count']/5000
```

```python
links = []
for i in range(len(od1)):
    r = od1.iloc[i]
    links.append({'source':r[0],'target':r[1],'value':float(r[2]),'lineStyle':{'width':float(r['width'])}})
links
```

```python
len(links)
```

```python
nodeinfo = od.groupby(['居住地_省','居住地_市'])['count'].sum().reset_index()
```

```python
nodeinfo['size'] = nodeinfo['count']**0.5/5
nodeinfo = nodeinfo.sort_values(by = ['居住地_省','count'],ascending=False)
```

```python
nodes = []
for i in range(len(nodeinfo)):
    r = nodeinfo.iloc[i]
    nodes.append({'name':r[1],
    'category':r[0],
    'value':float(r[2]),'symbolSize':float(r['size'])})
nodes
```

```python
len(nodes)
```

```python
len(links)
```

```python
data= [links,nodes]
```
